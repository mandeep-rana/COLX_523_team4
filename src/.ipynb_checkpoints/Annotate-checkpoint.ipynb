```python
import pandas as pd
from collections import defaultdict
import nltk
import json
from nltk.metrics.agreement import AnnotationTask
```

```python
annotated=pd.read_csv('sampled_reviews_annotated.csv')
```

```python
annotated=annotated.to_dict('records')
```

```python
annotated

each_cat=('rev','worker','y/n')
```

```python
#annotate_group_by_HITId=defaultdict(lambda: defaultdict(list))
blank_annotation={'Content':defaultdict(lambda: defaultdict(list)),
                  'Course delivery':defaultdict(lambda: defaultdict(list)),
                  'Difficulty':defaultdict(lambda: defaultdict(list)),
                  'Workload':defaultdict(lambda: defaultdict(list)),
                  'Usefulness':defaultdict(lambda: defaultdict(list))}
for i in annotated:
    identified_facets=json.loads(i['Answer.taskAnswers'])[0]['category']['labels']
    for facet in blank_annotation.keys():#initialize all to False by default
        blank_annotation[facet][i['WorkerId']][i['HITId']]='F'
    for identified in identified_facets:
        blank_annotation[identified][i['WorkerId']][i['HITId']]='T'
    #annotate_group_by_HITId[i['HITId']][i['WorkerId']]=blank_annotation
```

```python
def convert2triplets(annotation_of_HITId):
    trip=[]
    for worker in annotation_of_HITId.keys():
        for HITId in annotation_of_HITId[worker].keys():
            trip.append((worker,HITId,annotation_of_HITId[worker][HITId]))
    return trip
```

```python
for Facet in blank_annotation.keys():
    trip=convert2triplets(blank_annotation[Facet])
    trip=[i for i in trip if i[0] in ['A2G44A4ZPWRPXU','A61LR9E9L004M','AO2WNSGOXAX52']]
    annotation_task = AnnotationTask(trip)
    print(annotation_task)
    break
```
